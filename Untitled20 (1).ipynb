## FEATURE ENGINEERING

We will create new columns to determine the effectiveness of our delivery efficiency
1. THe delivery lead days is the difference between the number of days the product was ordered to when it was delivered
2. The delivery_efficiency is to determine if the product was delivered before, during or after the expected date.
3. The efficient delivery column is a binary column signaling if the product was delivered before or on the expected date or after the expected date. 1 is for before or on expected date, 0 is for after the expected date.

```python
df_ml_tr['Delivery_efficiency'] = (df_ml_tr['Delivered Date'] - df_ml_tr['Planned Date']).dt.days
df_ml_tr['Delivery_lead_days'] = (df_ml_tr['Delivered Date'] - df_ml_tr['Date']).dt.days
```

```python
import numpy as np

df_ml_tr['efficient_delivery'] = np.where(df_ml_tr['Delivery_efficiency'] <= 0, 1, 0)
```

```python
#Using Chi-square

categorical_columns = ['ID','Entry ID','TYPE', 'Destination', 'Group', 'Product', 'Order Source', 'Logistic Partner', 
                       'Deposit', 'Quarter', 'ABC','Customer Group']

chi_square_results = []

for col in categorical_columns:
    if col != 'TYPE': 
        contingency_table = pd.crosstab(df_ml_tr['ABC'], df_ml_tr[col])
        chi2, p, _, _ = chi2_contingency(contingency_table)
        chi_square_results.append({'Variable': col, 'Chi-square': chi2, 'P-value': p})

chi_square_results_df = pd.DataFrame(chi_square_results)

print("Chi-square test results:")
print(chi_square_results_df)
```
